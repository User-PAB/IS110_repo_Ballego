```python
total = 0
count = 0

while True:
    sval = input("Enter a number: ")
    if sval == "done":
        break
    try:
        fval = float(sval)
    except:
        print("Invalid input")
        continue

    total = total + fval
    count = count + 1

print(total, count, total / count)
```

```python
largest = None
smallest = None

while True:
    num = input("Enter a number: ")
    if num == "done":
        break
    try:
        fnum = float(num)
    except:
        print("Invalid input")
        continue

    # Update largest and smallest
    if largest is None or fnum > largest:
        largest = fnum
    if smallest is None or fnum < smallest:
        smallest = fnum

print("Maximum is", largest)
print("Minimum is", smallest)
```
