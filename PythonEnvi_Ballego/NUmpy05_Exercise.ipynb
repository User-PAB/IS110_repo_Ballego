```python
#A
nested_list = [[10, 20, 30], [40, 50, 60], [70, 80, 90]]   # make a nested list (like a table)

A = np.array(nested_list)   # turn the list into a NumPy array

print(A.shape)   # property: shows rowss and columns → (3, 3)

print(A.max())   # method: finds the biggest nuMber → 90

print(np.max(A)) # function: NumPy finds the biggest number → 90
```

```python
#B
B = np.ones((5, 5), dtype=int)   # make a 5x5 table filled with 1s (integers)

print(B.shape)   # shows shape → (5, 5)

print(B.size)    # shows total number of elements → 25

print(B.ndim)    # shows number of dimensions → 2 (rows and columns)
```

```python
#C
C = np.arange(-5, 6)   # make numbers from -5 up to +5

C_multiplied = C * 2   # multiply each number by 2

print(C_multiplied)    # print the result
```
