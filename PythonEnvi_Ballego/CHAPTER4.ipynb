```python
hours = float(input("Enter Hours: "))
rate = float(input("Enter Rate: "))
def computepay(hours, rate):
  if hours <= 40:
    pay = hours * rate
  else:
    regular = 40 * rate
    overtime = (hours - 40) * rate * 1.5
    pay = regular + overtime
    return pay
p = computepay(hours, rate)
print("Pay:", p)
```

```python
def computegrade(score):
    """
    Compute the grade based on the score using the given table.
    Returns the grade as a string.
    """
    if score < 0 or score > 100:
        return "Bad Score"

    if 97 <= score <= 100:
        return "1.0"
    elif 93 <= score < 97:
        return "1.25"
    elif 89 <= score < 93:
        return "1.5"
    elif 85 <= score < 89:
        return "1.75"
    elif 81 <= score < 85:
        return "2.0"
    elif 77 <= score < 81:
        return "2.25"
    elif 74 <= score < 77:
        return "2.5"
    else:  # 0 <= score < 74
        return "3.0"

score = float(input("Enter score: "))
grade = computegrade(score)
print("Grade", grade)
```
