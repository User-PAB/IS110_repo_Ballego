```python
df = pd.read_csv('/content/drive/MyDrive/IS110_BPerformance_Ballego/IS110_repo_Ballego/Dashboard_Ballego/asset/data/Best_Games_of_All_Time.csv')

df.head()
```

```python
Scatter plot + rolling average - Metascore
```

```python
import plotly.express as px
import plotly.graph_objects as go

df['Launch_date'] = pd.to_datetime(df['Launch_date'])
df = df.sort_values('Launch_date')

fig = px.scatter(df, x='Launch_date', y='Metascore',
                 color='Rating', hover_data=['Name'],
                 title='Metascore vs Launch Date')

df['Rolling'] = df['Metascore'].rolling(10).mean()
fig.add_trace(go.Scatter(x=df['Launch_date'], y=df['Rolling'],
                         mode='lines', name='Rolling Avg',
                         line=dict(color='blue', width=3)))

fig.update_layout(plot_bgcolor='white',
                  xaxis_title='Launch Date',
                  yaxis_title='Metascore')
fig.show()
```

```python
top_games = df.sort_values('Metascore', ascending=False).head(10)
fig = px.bar(top_games, x='Name', y='Metascore',
             title='Top 10 Games by Metascore',
             color='Metascore', color_continuous_scale='Blues',
             text_auto=True)
fig.update_layout(plot_bgcolor='white')
fig.show()
```

```python
Top 10 games ranked by Metascore each Decade
```

```python
Pie chart - Distribution By ESRB Rating
```

```python
 fig = px.pie(df, names='Rating', title='Games by ESRB Rating',
             color_discrete_sequence=px.colors.sequential.Blues_r)
fig.update_traces(textposition='outside', textinfo='percent+label')
fig.show()
```

```python
Histogram - Frequency Of Metascore Ratings
```

```python
fig = px.histogram(df, x='Metascore', nbins=10,
                   title='Distribution of Metascores',
                   color_discrete_sequence=['#1F4E79'])
fig.show()
```

```python
Box plot - Metascore Spread by  ESRB rating
```

```python
fig = px.box(df, x='Rating', y='Metascore',
             title='Metascore Spread by ESRB Rating',
             color='Rating', color_discrete_sequence=px.colors.qualitative.Set2)
fig.show()
```
